```python
df_publicadas = df_publicadas.drop(903)
```

```python
df_publicadas.isna().sum()
```

```python
df_provincia =df_publicadas.groupby(['Provincia','Cantón',
                                     'Entidad contratante']).agg(Suma_montos =("Monto contratado",'sum'))
df_provincia.sort_values("Provincia",ascending=True)
```

```python
#Vemos los totales por provinvia
df_topProvincia = df_provincia.groupby("Provincia").sum("Suma_montos").sort_values("Suma_montos",ascending=False)

df_topProvincia
```

```python
#sacamos el top 5 de provincias
df_topProvincia5 = df_topProvincia.sort_values("Suma_montos",ascending=False).head(5)
df_topProvincia5.sort_values("Suma_montos",ascending=False).head(5)
```

```python
#Seaborn barplot
sns.set_style("whitegrid")
bar,ax = plt.subplots(figsize=(10,6))
ax = sns.barplot(x=df_topProvincia5.index, y='Suma_montos', data=df_topProvincia5, ci=None, palette="muted",orient='v', )
ax.set_title("Top 5 de Provincias que más contrataron en Emergencia", fontsize=15)
ax.set_xlabel ("")
ax.set_ylabel ("Monto Contratado")

# Etiqueto 
for p in ax.patches:
 # get the height of each bar
 height = round(p.get_height(),2) 
 # adding text to each bar
 ax.text(x = p.get_x()+(p.get_width()/2), y = height+100,
 s = '{:,}'.format(height), ha = 'center')
```
